```python
import numpy as np
from sklearn.naive_bayes import (
  BernoulliNB,
  MultinomialNB 
)
```

### Bernoulli Naive Bayes vs. Multinomial Naive Bayes

Aunque ambos algoritmos pueden manejar datos binarios, sus suposiciones subyacentes son distintas, lo que explica las diferencias en las métricas. 

**Diferencia Clave**: MultinomialNB trata los 0/1 como frecuencias, mientras que BernoulliNB los trata como indicadores booleanos (presencia/ausencia)
- *MultinomialNB*: Calcula probabilidades basándose en la proporción de cada característica respecto al total de frecuencias de la clase. Un "1" cuenta como una ocurrencia
- *BernoulliNB*: Calcula probabilidades basándose en la proporción de muestras donde la característica está presente en la clase. Solo considera si aparece o no 


```python
# Datos binarios: [palabra_A, palabra_B, palabra_C]
X = [[1, 0, 1], [0, 1, 1], [1, 1, 0], [0, 0, 1]]
y = ['clase1', 'clase1', 'clase2', 'clase2']

# Multinomial trata los 1s como conteos
multinomial = MultinomialNB(alpha=1)
multinomial.fit(X, y)

# Bernoulli trata como verdadero booleano
bernoulli = BernoulliNB(alpha=1)
bernoulli.fit(X, y)

# Misma predicción pero probabilidades diferentes
new_value = [[1, 0, 0]]
print("Multinomial:", multinomial.predict(new_value))
print("Bernoulli:", bernoulli.predict(new_value))
print("Prob Multinomial:", multinomial.predict_proba(new_value))
print("Prob Bernoulli:", bernoulli.predict_proba(new_value))
```

**Resultados Diferentes**: Aún con los mismos datos binarios, las predicciones pueden variar $\implies$ **Causa**: Las fórmulas de verosimilitud son distintas $\to$ Estimaciones de probabilidad diferente
