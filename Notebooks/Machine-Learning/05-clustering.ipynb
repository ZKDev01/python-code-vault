```python
# modules
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN 

# path
HOUSE_PRICES_TRAIN = 'database/house-prices-train.csv'
```

# Clustering 

## Métricas de Distancia 

- Distancia Euclidiana
- Distancia Chebyshev
- Distancia Manhattan


```python
data = pd.read_csv (HOUSE_PRICES_TRAIN)
data
```

```python
data = data[[
  'Id',  
  'LotFrontage', 
  'LotArea', 
  'SalePrice'
]]
data.info()
```

```python
distances = [ 'euclidean', 'chebyshev', 'manhattan' ]

fig = plt.figure (figsize=(12,4))
for i,d in enumerate(distances):
  clusters = DBSCAN (2,min_samples=10,metric=d).fit_predict (data)
  ax = fig.add_subplot(1,3,i+1)
  ax.scatter (data[clusters==1][:,0],data[clusters==-1][:,1],c='lightgray',s=200,marker="2")
  
```
