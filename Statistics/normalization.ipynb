Usaremos la librería de Pandas de Python para poder codificar y obtener tres tipos de normalización.

- Normalización de z-score
- La normalización Min-Max
- Escalado simple

```python
import pandas as pd 
```

```python
columns = [ 'Longitud', 'Ancho', 'Altura' ]
values = [ 
	[ 168.8,  64.1,  48.8 ],
	[ 168.8,  64.1,  48.8 ],  
	[ 172.2,  65.5,  52.4 ],  
	[ 176.0,  66.2,  54.3 ],  
	[ 176.0,  66.2,  54.3 ],  
	[ 177.3,  66.3,  53.1 ],  
	[ 192.7,  71.4,  55.7 ],  
	[ 192.7,  71.4,  55.7 ],  
	[ 192.7,  71.4,  55.9 ] ]

df = pd.DataFrame ( values, columns = columns )
df
```

```python
df [ 'Longitud. Z-score' ] = ( df[ 'Longitud' ] - df[ 'Longitud' ].mean() ) / df[ 'Longitud' ].std()
df
```

```python
df [ 'Longitud. Min-Max' ] = ( df[ 'Longitud' ] - df[ 'Longitud' ].min() ) / ( df[ 'Longitud' ].max() - df[ 'Longitud' ].min() )
df
```

```python
df [ 'Longitud. Simple' ] = df[ 'Longitud' ] / df[ 'Longitud' ].max()
df
```
